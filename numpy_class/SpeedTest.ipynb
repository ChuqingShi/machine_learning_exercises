```python
from timeit import timeit

import numpy as np

def test1_slow(a,b):
    dot1 = 0
    for e,f in zip(a,b):
        dot1 += e*f
    return dot1

def test2_slow(a,b):
    dot2 = 0
    for i in range(len(a)):
        dot2 += a[i]*b[i]
    return dot2

def test3_slow(a,b):
    return np.array([e*f for e,f in zip(a,b)]).sum()

a = np.random.randn(100)
b = np.random.randn(100)
T = 100000

#use timeit with lambda function
#lambda: allows to call the function without calling it directly before passing it to timeit().
time1_slow = timeit(lambda:test1_slow(a,b), number=T)
time2_slow = timeit(lambda:test2_slow(a,b), number=T)
time3_slow = timeit(lambda:test3_slow(a,b), number=T)
time1_np = timeit(lambda: np.dot(a,b), number=T)
time2_np = timeit(lambda: a.dot(b), number=T)
time3_np = timeit(lambda: a@b, number=T)

print(f"{time1_slow/time1_np : .5f} time faster.") #for with zip
print(f"{time2_slow/time1_np : .5f} time faster.") #for with index
print(f"{time3_slow/time1_np : .5f} time faster.") #list comprehension
print(f"{time1_slow/time2_np : .5f} time faster.") #for with zip
print(f"{time2_slow/time2_np : .5f} time faster.") #for with index
print(f"{time3_slow/time2_np : .5f} time faster.") #list comprehension
print(f"{time1_slow/time3_np : .5f} time faster.") #for with zip
print(f"{time2_slow/time3_np : .5f} time faster.") #for with index
print(f"{time3_slow/time3_np : .5f} time faster.") #list comprehension
```

## conclusion:
* speed a.dot(b) > np.dot(a,b) > a@b
* speed ziploop > comprehensionloop > indexloop

### Best use a.dot(b) for dot product.
### Best use item zip in for loop.
